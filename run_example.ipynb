```python
import numpy as np
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

import easyocr # https://github.com/JaidedAI/EasyOCR
```

## Initial reader AI model

```python
reader = easyocr.Reader(['en']) # Read English (en), for other language: https://www.jaided.ai/easyocr/
```

## Read simple image

```python
image = Image.open('images/banghieu2.jpg').convert("RGB")
image
```

## OCR

```python
result = reader.readtext(np.array(image))
```

```python
result
```

## Visualize

```python
drawer = ImageDraw.Draw(image)

for text_polygon, text, probability in result:
    text_polygon_list_of_tuple = [tuple(point) for point in text_polygon]
    print(text_polygon_list_of_tuple, text, probability)
    drawer.polygon(text_polygon_list_of_tuple, outline='yellow')
    drawer.text(text_polygon_list_of_tuple[0], text, fill='yellow')

image
```
